### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv). 

### Step 3. Assign it to a variable called chipo.

```python
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv'
chipo = pd.read_csv(url, sep = '\t')
chipo.head(10)
```

### Step 4. How many products cost more than $10.00?

```python
# change the item_price to numeric values
chipo['item_price'] = chipo.apply(lambda x: float(x['item_price'].replace('$', '')), axis = 1)
chipo.head(10)
```

```python
chipo.loc[chipo.item_price > 10.00]['item_name'].nunique()
```

### Step 5. What is the price of each item? 
###### print a data frame with only two columns item_name and item_price

```python
# unique combinations
chipo.loc[chipo.item_price > 10.00][['item_name', 'item_price']].drop_duplicates().reset_index(drop=True)
```

### Step 6. Sort by the name of the item

```python
chipo.loc[chipo.item_price > 10.00][['item_name', 'item_price']].drop_duplicates().sort_values(['item_name']).reset_index(drop=True)
```

### Step 7. What was the quantity of the most expensive item ordered?

```python
chipo.iloc[chipo['item_price'].idxmax()]['quantity']
```

### Step 8. How many times was a Veggie Salad Bowl ordered?

```python
chipo.loc[chipo.item_name == 'Veggie Salad Bowl']['order_id'].nunique()
```

### Step 9. How many times did someone order more than one Canned Soda?

```python
chipo.loc[(chipo.item_name == 'Canned Soda') & (chipo.quantity > 1)]['order_id'].nunique()
```
